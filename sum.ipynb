```python
#安全库存函数

import pandas as pd
import numpy as np
import scipy.stats as stats
import math


def safety_stock_function(file_path, service_level):
    # 读取xlsx表格中的数据
    df = pd.read_excel(file_path)

    # 选择时间和需求量数据
    time = df.iloc[:, 0]  # 第一列是时间
    demand = df.iloc[:, 2]  # 第三列是需求量

    # 计算需求量的平均值和标准差
    mean_demand = demand.mean()
    #print(mean_demand)
    std_demand = demand.std()

    # 计算z值（正态分布的分位数）
    z_value = stats.norm.ppf(service_level)

    # 计算安全库存
    safety_stock = z_value * std_demand
    
    return math.ceil(safety_stock)

print(safety_stock_function('./6004020900.xlsx',0.85))
```

```python
#加上安全库存+实际生产量

#满意度 = 1-（第i周的预测值 - 第i周的真实值）/第i周的真实值

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import math 

# 设置字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 读取时间序列数据
file_path = './6004100008.xlsx'
df = pd.read_excel(file_path)

#计算安全库存
safety_stock = safety_stock_function(file_path , 0.85)

# 选择时间序列数据
time_series = df.iloc[:, 2]  # 第三列是数据
time_series.index = pd.RangeIndex(start=1, stop=len(time_series) + 1, step=1)

# 划分数据集，70% 作为训练集，30% 作为验证集
train_size =100
train, test = time_series[:train_size], time_series[train_size:]

p = 1
d = 1
q = 0

# 设置 ARIMA 模型的最佳参数 (假设已知)
best_order = (p, d, q)  # 替换为实际的最优参数

# 存储预测值的列表
predictions = []

# 存储误差的列表
errors = []

#计算满意度，积压量，缺货量，实际需求量
actual_prodction = 0  #实际生产量
actual_need = 0           #实际需求量
need = []
actual_pro = []              
average_num = 0        #用来计算平均满意度的
num = 0
last_pred = 0                #
temp_surplus = 0        #积压量
temp_miss = 0             #缺货量
surplus = []
miss = []
degree_sa = 0 
degree = []                   #满意度


# 循环预测每一个时间点
for i in range(len(test)):
    # 拟合 ARIMA 模型
    model = ARIMA(train, order=best_order)
    model_fit = model.fit()

    # 预测下一个时间点的值
    next_pred = model_fit.forecast(steps=1).iloc[0]  # 使用 iloc[0] 获取预测值
    next_pred = math.ceil(next_pred)  #向上取整
    predictions.append(next_pred)

    # 获取原本的需求
    actual_value = test.iloc[i]
    #error = abs((next_pred - actual_value) / actual_value)
     
    #判断实际生产量是否会是负数
    #if next_pred + safety_stock - temp_surplus + temp_miss < 0:
        #actual_production = 0
    #else :
        #actual_production = next_pred + safety_stock - temp_surplus + temp_miss
    
    if temp_surplus > next_pred + safety_stock + temp_miss:
        actual_production = 0
        actual_need = actual_value + temp_miss
        
        if actual_production + temp_surplus >actual_need :
            temp_surplus = actual_production + temp_surplus -actual_need
            temp_miss = 0
            degree_sa = 1 
        else :
            temp_miss =  actual_need - actual_production - temp_surplus 
            temp_surplus = 0
            degree_sa = (1 - temp_miss/actual_need)
            
    else :
            actual_production = temp_miss + next_pred + safety_stock - temp_surplus
            actual_need = temp_miss + actual_value
            
            if actual_production + temp_surplus > actual_need :
                temp_surplus = actual_production + temp_surplus - actual_need
                temp_miss = 0
                degree_sa = 1
            else :
                temp_miss =  actual_need - actual_production - temp_surplus
                temp_surplus = 0
                degree_sa = (1 - temp_miss/actual_need)
    
    surplus.append(temp_surplus)
    miss.append(temp_miss)
    actual_pro.append(actual_production)
    #errors.append(error)
    degree.append(degree_sa)
    need.append(actual_need)

    # 将下一个时间点的实际值加入到训练集中，重新拟合
    new_index = test.index[i]  # 获取当前预测时间点的索引
    train = pd.concat([train, pd.Series(actual_value, index=[new_index])])

# 将预测值转换为时间序列格式
pred_series = pd.Series(predictions, index=test.index)

# 将误差转换为时间序列格式
#error_series = pd.Series(errors, index=test.index)

#多余的
surplus_series = pd.Series(surplus, index=test.index)

#缺少的
miss_series = pd.Series(miss, index=test.index)

#实际生产的
actual_pro_series = pd.Series(actual_pro, index=test.index)

#满意度
degree_series = pd.Series(degree, index=test.index)

#实际的需求（需求+缺货）
need_series = pd.Series(need, index=test.index)

# 周数（时间）从101开始
weeks = list(range(101, 101 + len(actual_pro)))

# 创建DataFrame
output_df = pd.DataFrame({
    '周数': weeks,
    '生产计划/件': actual_pro,
    '实际需求量/件': need,
    '库存量/件': surplus,
    '缺货量/件': miss,
    '服务水平': degree
})

# 将DataFrame保存为xlsx文件
output_file_path = './output_data.xlsx'  # 设置输出文件路径
output_df.to_excel(output_file_path, index=False)

print(f"数据已保存到 {output_file_path}")

# 绘制实际值与预测值的对比图
plt.figure(figsize=(12, 8))
plt.plot(time_series.index, time_series, label='实际数据')
plt.plot(pred_series.index, pred_series, label='逐步预测值', color='red')
plt.xlabel('时间')
plt.ylabel('数值')
plt.title('逐步ARIMA预测')
plt.legend()
plt.show()

# 绘制误差图
#plt.figure(figsize=(12, 8))
#plt.plot(error_series.index, error_series, label='预测误差', color='blue')
#plt.xlabel('时间')
#plt.ylabel('误差值')
#plt.title('预测误差随时间变化')
#plt.legend()
#plt.show()

# 绘图
plt.figure(figsize=(12, 8))
plt.plot(miss_series.index, miss_series, label='缺少货物量', color='blue')
plt.xlabel('时间')
plt.ylabel('缺少量')
plt.title('缺少货物量随时间变化')
plt.legend()
plt.show()

# 绘图
plt.figure(figsize=(12, 8))
plt.plot(surplus_series.index, surplus_series, label='多余货物量', color='blue')
plt.xlabel('时间')
plt.ylabel('多余货物量')
plt.title('多余货物量随时间变化')
plt.legend()
plt.show()

# 绘图
plt.figure(figsize=(12, 8))
plt.plot(actual_pro_series.index, actual_pro_series, label='实际生产货物量', color='blue')
plt.xlabel('时间')
plt.ylabel('实际生产货物量')
plt.title('实际生产货物量随时间变化')
plt.legend()
plt.show()

# 绘图
plt.figure(figsize=(12, 8))
plt.plot(degree_series.index, degree_series, label='满意度', color='blue')
plt.xlabel('时间')
plt.ylabel('满意度')
plt.title('满意度随时间变化')
plt.legend()
plt.show()

# 绘图
plt.figure(figsize=(12, 8))
plt.plot(need_series.index, need_series, label='实际需求量', color='blue')
plt.xlabel('时间')
plt.ylabel('实际需求量')
plt.title('实际需求量随时间变化')
plt.legend()
plt.show()

print(1-average_num/len(test))
```
