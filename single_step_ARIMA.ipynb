```python
#ACF/PACF

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
file_path = './updated_file.xlsx'
df = pd.read_excel(file_path)

# 
time_series = df.iloc[:, 2]  # 
time_series.index = pd.RangeIndex(start=1, stop=len(time_series) + 1, step=1)

# 70% 30% 
train_size = int(len(time_series) * 0.7)
train, test = time_series[:train_size], time_series[train_size:]

p = 1
d = 1
q = 0

#  ARIMA  ()
best_order = (p, d, q)

# 
predictions = []
residuals = []

# 
for i in range(len(test)):
    #  ARIMA 
    model = ARIMA(train, order=best_order)
    model_fit = model.fit()

    # 
    next_pred = model_fit.forecast(steps=1).iloc[0]
    predictions.append(next_pred)

    # 
    actual_value = test.iloc[i]
    residual = actual_value - next_pred
    residuals.append(residual)

    # 
    new_index = test.index[i]
    train = pd.concat([train, pd.Series(actual_value, index=[new_index])])

# 
pred_series = pd.Series(predictions, index=test.index)
residual_series = pd.Series(residuals, index=test.index)

# 
plt.figure(figsize=(12, 8))
plt.plot(time_series.index, time_series, label='')
plt.plot(pred_series.index, pred_series, label='', color='red')
plt.xlabel('')
plt.ylabel('')
plt.title('ARIMA')
plt.legend()
plt.show()

# 
plt.figure(figsize=(12, 8))
plt.plot(residual_series.index, residual_series, label='', color='green')
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.legend()
plt.show()

#  ACF 
plt.figure(figsize=(12, 8))
plot_acf(residual_series, lags=20)
plt.title('ACF')
plt.show()

#  PACF 
plt.figure(figsize=(12, 8))
plot_pacf(residual_series, lags=20)
plt.title('PACF')
plt.show()
```
